```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
average_ratings = ratings.groupby('movieId').mean()[['rating']].round(1).reset_index()
```

```python
def average_rating(rating_class):
    if rating_class['rating'] <= 2: 
        return 'низкий рейтинг'
    elif rating_class['rating'] <= 4: 
        return 'средний рейтинг'
    else: return 'высокий рейтинг'
```

```python
average_ratings['class'] = average_ratings.apply(average_rating, axis=1)
average_ratings.head()
```

```python
key = pd.read_csv('keywords.csv')
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def aa(geo_class):
    for key, values in geo_data.items():
        for i in values:
            if i in (geo_class['keyword']).lower():
                return key
    else: return 'undefined'
```

```python
key['geo'] = key.apply(aa, axis=1)
```

```python
key.sort_values(by='geo')
```

```python
import numpy as np
```

```python
years = np.arange(1950, 2010)
```

```python
def production_year(yea):
    for i in years:
        if str(i) in yea:
            return i
    return 1900
    
```

```python
movies['year'] = movies['title'].apply(production_year)
movies.head()
```

```python
join_ = pd.merge(movies, ratings, on='movieId', how='left')
```

```python
a = join_.groupby('year').agg({'rating': 'mean'}).round(1)
```

```python
a.sort_values(by='rating', ascending=False)
```
