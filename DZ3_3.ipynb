```python
log = pd.read_csv('visit_log.csv', sep = ';')
log
```

```python
def aa(log_):
    if (log_['traffic_source'] == 'yandex') | (log_['traffic_source'] == 'google'):
        return 'organic'
    elif (log_['traffic_source'] == 'paid') | (log_['region'] == 'Russia'):
        return 'ad'
    elif (log_['traffic_source'] == 'paid') & ~ (log_['region'] == 'Russia'):
        return 'other'
    else:
        return log_['traffic_source']
```

```python
log['sourse_type'] = log.apply(aa, axis = 1)
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings_count = ratings.groupby('userId').count().reset_index()
ratings_count = ratings_count[ratings_count['rating'] > 100]
```

```python
ratings_count[['userId','rating']]
```

```python
ratings_max = (ratings.groupby('userId').max().reset_index())[['userId','timestamp']]
ratings_min = (ratings.groupby('userId').min().reset_index())[['userId','timestamp']]
```

```python
ratings_max.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
ratings_min.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
```

```python
full_rating = ratings_count.merge(ratings_min, how='left', on='userId').merge(ratings_max, how='left', on='userId')
full_rating.head()
```

```python
def avg(full_rat):
    a = ((full_rat['timestamp_max']) - (full_rat['timestamp_min']))
    return  pd.to_datetime(a, unit='s')
```

```python
full_rating['avg'] = full_rating.apply(avg, axis = 1)
full_rating.head()
```
